Price Tracker (ograniczam sie do kategorii URODA)

Co potrzebujemy zrobic/dowiedziec sie:
- z jakich stron chcemy robic web scrapping 
- jak zrobic web scrapping 
- jakie elementy chce pokazywac w moim trackerze: 
    - User musi miec mozliwosc dodawania, edytowania i usuwania produktow ktore chce trackowac
    - Kazdy product entry powinien zawierac informacje o nazwie produktu, URL strony, target price (opcjonalnie), krotki opis produktu
    - program powinien regularnie sprawdzac wskazany/wskazane linki URL pod kątem zmiany ceny produktu (częstotliwość sprawdzania powinna być mozliwa do ustawienia np. co godzine, raz dziennie itd)
    - program powinien zapisywac cene kazdego produktu co, jesli cena zmieni sie w porownaniu z poprzednim sprawdzeniem, user powinien zostac poinformowany (przez wiadomosc email, powiadomienie na desktopie czy wiadomosc w terminalu)
    - chcemy moc pokazac historie cen  dla kazdego produktu (z uwzglednieniem daty/czasu i ceny)
    - user powinien moc wyswietlic sobie wszystkie produkty ktore dodal do trackowania, filtrowac je po spadku ceny (od najwyzszego spadku do najmniejszego i na odwrot), po wzroscie ceny, wyswietlanie tylko produktow ktore sa ponizej target price 
    - statystyki typu liczba spadkow cen w ostatnim tygodniu, srednia zmiana ceny per produkt, produkty ktore aktualnie sa w najnizszej zanotowanej cenie 

Jeśli chodzi o dane ktore bedzie wykorzystywal tracker (lista produktow, historia cen, user settings) mogą być one przechowywane w pliku CSV albo w bazie danych SQL.
Do testowania chcemy stworzyc dummy-product-list (z conajmniej 3 produktami) i funkcje ktora dodaje dane o tych produktach do bazy danych aplikacji/pliku CSV. 
Program powinien uwzględniac mozliwosc wystepowania bledow takich jak niepoprawne URL, niedostepne strony, brakujace dane o produkcie 
User interface moze byc terminal based albo w postaci prostego GUI. 

Mini architektura:
- database.py -> komunikacja z bazą, przechowywanie danych
- scraper.py -> pobieranie ceny 
- tracker.py -> logika decyzyjna 
- notifications.py -> wysyłanie alertów 
- main.py -> uruchamianie
